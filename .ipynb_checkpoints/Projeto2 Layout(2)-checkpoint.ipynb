## 1. Problema

O Classificador Naive-Bayes, o qual se baseia no uso do teorema de Bayes, é largamente utilizado em filtros anti-spam de e-mails. O classificador permite calcular qual a probabilidade de uma mensagem ser SPAM considerando as palavras em seu conteúdo e, de forma complementar, permite calcular a probabilidade de uma mensagem ser HAM dada as palavras descritas na mensagem.

Para realizar o MVP (minimum viable product) do projeto, você precisa programar uma versão do classificador que "aprende" o que é uma mensagem SPAM considerando uma base de treinamento e comparar o desempenho dos resultados com uma base de testes.

## 2. Separação da base de dados em Treinamento e Teste

A base de dados deve ser separada em duas partes, aleatoriamente, considerando: 
    
    75% dos dados para a parte Treinamento; e
    25% dos dados para a parte Teste.

```python
import pandas as pd
import os
import random
import numpy as np
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
leitura = pd.ExcelFile('spamham2019(1).xlsx')
dados = pd.read_excel(leitura)
dados.head()
```

```python
#Separar tabela em treino e teste:
indices = range(len(dados))
indTreino = random.sample(indices, int(len(indices)*0.75))
indTeste = []
for i in indices:
    if i not in indTreino:
        indTeste.append(i)

treino = dados.loc[indTreino,:]
teste = dados.loc[indTeste,:]

#Cria índices numéricos pros dados
treino.index = range(len(treino))
teste.index = range(len(teste))
```

# Limpando a base de dados

```python
tirar = ['!','@','#','$','%','¨','&','*','(',')','-','_','+','=','§',',','.','<','>',':',';','/','?','°','|','\\','{','}','ª','[',']','º','×','–','—','•','…','‹','›','€','1','2','3','4','5','6','7','8','9','0']
for i in tirar:
    treino.Email = treino.Email.str.replace(i, '', regex=False)
    teste.Email = teste.Email.str.replace(i, '', regex=False)
```

```python
#CRIAR TABELA DE FREQUENCIA DE CADA PALAVRA
freq = {} #Tabela de frequências | 'R' = cont. Relevante, 'I' cont Irrelevante, 'pR' = prob palavra aparecer dado que o email é relevante

for numLinha in range(len(treino.Email)): 
    palavras_linha = treino.loc[numLinha,'Email'].split(' ')
    for palavra in palavras_linha:
        if palavra not in freq:
            freq[palavra] = {'R':0,'I':0,'pPR':0,'pPI':0}
        if treino.loc[numLinha,'Class'] == 'ham':
            freq[palavra]['R'] +=1
        else:
            freq[palavra]['I'] +=1
```

```python
totalR,totalI = 0,0 #total de palavras relevantes e  total de palavras irrelevantes (contando repetições)

PR = len(treino[treino['Class']=='ham'])/len(treino) #prob email ser relevante 

for key,valor in freq.items():  
    totalR += valor['R']
    totalI += valor['I']
        
for palavra,valor in freq.items():
    pP = (valor['R']+valor['I'])/(totalR+totalI)
    freq[palavra]['pPR'] = (valor['R']+1)/(PR*(totalR+totalI+1))#(2*totalR+totalI)
    freq[palavra]['pPI'] = (valor['I']+1)/(PR*(totalR+totalI+1))#/(2*totalI+totalR)
```

## 3. Classificador Naive-Bayes

```python
for numLinha in range(len(teste.Email)): # 'R' = Relevante e 'I' Irrelevante
    palavras_linha = teste.loc[numLinha,'Email'].split(' ')
    for j in palavras_linha:
        
        
        if j not in freq:
            if teste.loc[numLinha,'Class'] == 'ham':
                freq[j] = {'R':1,'I':0}
            else:
                freq[j] = {'R':0,'I':1}
        else:
            if teste.loc[numLinha,'Class'] == 'ham':
                freq[j]['R'] +=1
            else:
                freq[j]['I'] +=1
```
